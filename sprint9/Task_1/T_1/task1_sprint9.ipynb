 # Detect the edges of an image Sobel ,Lablacian and Canny

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('girl.png')
plt.imshow(image)
plt.show()
```

```python
lap=cv.Laplacian(image,cv.CV_64F,ksize=3)
plt.imshow(lap)
plt.show()
```

```python
sobelx = cv2.Sobel(image,cv.CV_64F,1,0,ksize=5)
#sobelx2=np.uint8(np.absolute(sobelx))
plt.imshow(sobelx)
plt.show()
```

```python
sobely = cv.Sobel(image,cv.CV_64F,0,1,ksize=5)
plt.imshow(sobely)
plt.show()
```

```python
combine_sobel=cv.bitwise_or(sobelx,sobely)
plt.imshow(combine_sobel)
plt.show()
```

```python
canny=cv.Canny(image,100,200)
plt.imshow(canny)
plt.show()
```

# Hough Line Transform

```python
img=cv.imread("lines_detected.jpg")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray,50,150,apertureSize=3)
lines = cv2.HoughLinesP(
            edges, # Input edge image
            1, # Distance resolution in pixels
            np.pi/180, # Angle resolution in radians
            threshold=100, # Min number of votes for valid line
            minLineLength=5, # Min allowed length of line
            maxLineGap=10 # Max allowed gap between line for joining them
            )
lines_list=[]
for points in lines:
      # Extracted points nested in the list
    x1,y1,x2,y2=points[0]
    # Draw the lines joing the points
    # On the original image
    cv2.line(img,(x1,y1),(x2,y2),(0,255,0),2)
    # Maintain a simples lookup list for points
    lines_list.append([(x1,y1),(x2,y2)])
     
plt.imshow(img)
plt.show()
```
